```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Ladda data
df = pd.read_csv('bank_data.csv')

# 2. Välj viktiga kolumner
X = df[['BALANCE', 'PURCHASES']].dropna()

# 3. Skala data (viktigt för bättre resultat)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Skapa K-means modell (3 kluster)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# 5. Lägg till cluster i datan
X['Cluster'] = kmeans.labels_

# 6. Visualisering (snygg och tydlig)
plt.figure(figsize=(8,6))
plt.scatter(X['BALANCE'], X['PURCHASES'], c=X['Cluster'])

plt.xlabel('Balance')
plt.ylabel('Purchases')
plt.title('Customer Segmentation (K-means)')

plt.show()

# 7. Visa hur många kunder i varje grupp
print(X['Cluster'].value_counts())
```
